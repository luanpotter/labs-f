```python
import numpy as np
import matplotlib.pyplot as plt

freq = .5
time = np.arange(1, 100, .001)
value_in_YUNits = [np.sin(2*np.pi*freq*t) for t in time]

#### fft ####
Ts = float(time[0])
n = len(time) # length of the signal
k = np.arange(n)
T = n*Ts
frq = k/T

fft = np.fft.fft(value_in_YUNits)/n # fft computing and normalization
fft_real = np.real(fft)

### gráfico ###
fig, ax = plt.subplots(2, 1)
ax[0].plot(time,value_in_YUNits)
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Amplitude')
ax[1].plot(time, fft_real)
ax[1].set_xlabel('Freq (Hz)')
ax[1].set_ylabel('|Y(freq)|')
ax[1].set_xlim(xmin=1)

plt.show()
```
